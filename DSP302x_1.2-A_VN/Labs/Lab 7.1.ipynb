# Lab: Làm việc với dữ liệu Object trong Pandas

# Đọc dữ liệu từ file

```python
import pandas as pd
file_path = r"D:\dangnh4\1. Tài liệu\0. Tài liệu bàn giao\Khóa học\Funix\Data\Google_Playstore.csv"


df_google_playstore = pd.read_csv(file_path)
df_google_playstore.head()
```

```python
# Yêu cầu: thay đổi gmail.com ở Developer Email thành gmails.com
df_google_playstore['Developer Email'].replace(to_replace=r"gmails.com$", value="gmail.com", regex=True, inplace=True)
df_google_playstore['Developer Email']
```

```python
# Chuyển đổi M trong cột Size thành MB, k trong cột Size thành kB
df_google_playstore['Size'].replace(to_replace="M$", value="MB", regex=True, inplace=True)
df_google_playstore['Size'].replace(to_replace="k$", value="kB", regex=True, inplace=True)

df_google_playstore['Size']
```

```python
# Yêu cầu: Lấy các giá trị trước dấu chấm "." đầu tiên trong App Id
df_google_playstore['App Id'].apply(lambda x: x.split(".")[0])
```

```python
# Yêu cầu: Quan sát các giá trị duy nhất và số lượng của giá trị trước dấu chấm "." đầu tiên trong App Id
print(df_google_playstore['App Id'].apply(lambda x : x.split(".")[0]))
print(df_google_playstore['App Id'].apply(lambda x : x.split(".")[0]).count())
```

```python
# Yêu cầu: Tạo một cột mới chứa đuôi của Email trong Developer Email
def tail_email(row: pd.Series):
    try:
        result = row.split("@")[-1]
    except AttributeError:
        result = None
    return result
     
df_google_playstore['Email Tail'] = df_google_playstore['Developer Email'].apply(tail_email)
df_google_playstore['Email Tail']
```
