```python
# Yêu cầu: Quan sát một số đặc điểm của trường dữ liệu dạng số
df_google_playstore...
```

```python
# Yêu cầu: Từ các cột dữ liệu dạng số, hãy quan sát:
# - Giá trị lớn nhất, 
# - Giá trị nhỏ nhất
# - Vị trí của giá trị lớn nhất
# - Vị trí của giá trị nhỏ nhất
# - Trung bình
# - Trung vị
# - Giá trị xuất hiện nhiều nhất
# - Tổng của toàn bộ dữ liệu
for col_name in df_google_playstore.select_dtypes(['float', 'int']).columns:
    print(col_name)
    print('Sum:', df_google_playstore[col_name]...)
    print('Min:', df_google_playstore[col_name]...)
    print('Max:', df_google_playstore[col_name]...)
    print('Index value min:', df_google_playstore[col_name]...)
    print('Index value max:', df_google_playstore[col_name]...)
    print('Mean:', df_google_playstore[col_name]...)
    print('Median:', df_google_playstore[col_name]...)
    print('Mode:', df_google_playstore[col_name]...)
    print()
```
